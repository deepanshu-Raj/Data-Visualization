# 2 . Scatter Plot

```python
#adjust the plot size.
plt.figure(figsize = (6,6))
plt.scatter(x,y,color = 'r',label = 'My_profit')
plt.scatter(x,y1,color = 'b',label = 'My_income')
plt.xlabel('time')
plt.ylabel('price')
plt.legend()
plt.show()
```

# 3 . Bar Graph

```python
plt.style.use('seaborn')
plt.figure(figsize = (8,8))
x_coord = np.array([1, 2, 3, 4])
plt.bar(x_coord-0.125,np.array([10,20,15,22]),width = 0.25,tick_label = ['gold','silver','bronze','platinum'],color = 'orange',label = 2019)
plt.bar(x_coord+0.125,np.array([22,15,20,10]),width = 0.25,tick_label = ['gold','silver','bronze','platinum'],color = 'green',label = 2020)
plt.xlabel('Metals')
plt.ylabel('Price')
plt.title('Metal Prices')
plt.ylim(0,30)
plt.legend()
plt.show()
```

# 4 . Pie Chart

```python
plt.figure(figsize = (8,8))
subjects = 'Mathematics','computer','english','physics','chemistry'
expl = (0.1,0,0,0.1,0)
vals = [95,95,89,90,95]
plt.pie(vals,labels = subjects,explode = expl,shadow = True,autopct = '%1.1f%%')
plt.title('Marks in 12th Board Exams')
plt.show()
```

# 5 . Histogram(Normal Distribution)

```python
#creating a normal distribution from standard normal distribution
X_standardNormal = np.random.randn(100)
sigma = 5
u = 75
X = np.round(sigma*X_standardNormal + u)
Y = np.round(sigma*X_standardNormal + u-20)
print(max(X),min(X))
#now, the values are spread in range [61,88]
#with mean = 75 ans std = 5
print(np.mean(X))
print(np.std(X))
```

```python
#take care of overlap , with alpha value in range (0,1).
plt.style.use('Solarize_Light2')
plt.hist(X,label = 'maths',color = 'yellow',alpha = 0.9)
plt.hist(Y,label = 'physics',color = 'orange',alpha = 0.5)
plt.ylim(0,25)
plt.xlim(20,100)
plt.title('Marks of the students')
plt.xlabel('marks')
plt.ylabel('freq. of number of students')
plt.legend()
plt.show()
```

# - Seaborn Usage

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head(10)
```

```python
print(tips.shape)
```

# 1 . Bar Plot

```python
#bargraph in seaborn
sns.barplot(x = 'sex',y = 'total_bill',data = tips,estimator = np.std)
plt.show()
```

# 2 . Count Plot

```python
#countPlot
plt.figure(figsize = (8,6))
sns.countplot(x = 'sex',data = tips)
plt.ylim(0,180)
plt.show()
```

```python
sns.countplot(x = 'time',data = tips)
plt.show()
```

# 3 . Box Plot

```python
#used with the categorical data.
#they show the distribution of categorical data.
plt.figure(figsize = (8,6))
sns.boxplot(x ='day' ,y ='total_bill' ,data = tips)
plt.show()
```

```python
#with hue : categorical data.
plt.figure(figsize = (8,6))
sns.boxplot(x ='day' ,y ='total_bill' ,data = tips,hue = 'sex')
plt.show()
```

# 4 . Violin Plot

```python
plt.style.use('seaborn')
sns.violinplot(x = 'day', y = 'total_bill',data = tips,hue = 'sex',split = True)
plt.show()
```

# 5 . Distribution plot & KDE plot & Joint Plot

```python
#kde : kernel density estimate
plt.figure(figsize = (6,4))
sns.distplot(tips['total_bill'],kde = True,color = 'orange')
plt.show()
plt.figure(figsize = (6,4))
sns.kdeplot(tips['total_bill'],color = 'green')
plt.show()
```

```python
#jointplot
sns.jointplot(x ='total_bill' ,y = 'size',data = tips,kind = 'hex',color = 'blue')
sns.jointplot(x= 'total_bill', y = 'tip',data = tips,kind = 'hex',color = 'blue')
plt.show()
```
